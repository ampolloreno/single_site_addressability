```julia
points = [collect(range(0+i*50, 50+i*50, step=1)) for i in range(0, 10, step=1)]
```

```julia
μs = [[i for p in point] for (i, point) in enumerate(points)]
```

```julia
using Plots
plot(points, μs, ylabel="nω", xlabel="μs", legend=false, title="Applied Beatnote Frequency versus Time")
```
