```julia
function dir(x)
    T = typeof(x)
    for (name, typ) in zip(fieldnames(T), T.types)
        println("type of the fieldname $name is $typ")
    end
end
function fidelity(ρ, σ)
    tr(sqrt(sqrt(ρ) * σ * sqrt(ρ)))^(1/2)
end
```

```julia
function R(n::Int64, m::Int64, ρ::Float64, scale::Float64=1E-6)
    """Following the naming convention on Wikipedia for R_n^m: https://en.wikipedia.org/wiki/Zernike_polynomials"""
    if (n - m) % 2 != 0
        0
    else
        function summand(k)
            if ρ == 0
                0
            else
                (-1)^k * factorial(n-k)/(factorial(k)*factorial(Int((n+m)/2) - k)*factorial(Int((n-m)/2) - k))*(ρ/scale)^(n-2*k)
            end
        end
        mapreduce(summand, +, Array(range(0, stop=Int((n-m)/2), step=1)))
    end
end

function Z(n::Int64, m::Int64)
    """In John's notes he includes a normalization, we have omitted that and are just following the conventions
    on Wikipedia."""
    function Z_at_point(ρ::Float64, ϕ::Float64)
        if m >= 0
            R(n, m, ρ) * cos(m * ϕ)
        else
            R(n, -m, ρ) * sin(-m * ϕ)
        end
    end
    Z_at_point
end
```

```julia
using SpecialFunctions
using QuantumOptics

function H_odf(ρ::Float64, ϕ::Float64, t::Float64, zernike_coefficients_even::Array,
        zernike_coefficients_odd::Array, U::Float64=2*π*10E3, ψ::Float64=0.;
        μ::Float64=0.0, ω::Float64=2*π*43.8E3, bessel=false, small=false)
    # For backwards compatibility - we might want different orders of Zernike terms.
    if !isa(zernike_coefficients_even[1], Array)
        zernike_coefficients_even = [zernike_coefficients_even]
    end
    if !isa(zernike_coefficients_odd[1], Array)
        zernike_coefficients_odd = [zernike_coefficients_odd]
    end
    
    eventuples = []
    for zeven in zernike_coefficients_even
        neven = length(zeven) - 1
        Peven = reduce(+, [zeven[i] * R(neven, i-1, ρ) for i in Array(range(1, neven + 1, step=1))])
        rotationeven = - (neven) * ω * t
        push!(eventuples, (neven, Peven, rotationeven))
    end
    oddtuples = []
    for zodd in zernike_coefficients_odd
        nodd = length(zodd) - 1
        Podd = reduce(+, [zodd[i] * R(nodd, i-1, ρ) for i in Array(range(1, nodd + 1, step=1))])
        rotationodd = - (nodd) * ω * t
        push!(oddtuples, (nodd, Podd, rotationodd))
    end
    
    # any bessel approximation should only apply one order of zernike, this should error otherwise
    if bessel && small
        U/2 * (mapreduce(x->x[2]*sin(x[1]*ϕ - ψ), +, eventuples)
            - mapreduce(x->x[2]*cos(x[1]*ϕ - ψ), +, oddtuples))
        #U/2 * (Peven * sin(neven * ϕ - ψ) - Podd * cos(nodd * ϕ - ψ))
    elseif bessel && !small
        U * (mapreduce(x->besselj1(x[2])*sin(x[1]*ϕ - ψ), +, eventuples)
            - mapreduce(x->besselj1(x[2])*cos(x[1]*ϕ - ψ), +, oddtuples))
        #U * (besselj1(Peven) * sin(neven * ϕ - ψ) - besselj1(Podd) * cos(nodd * ϕ - ψ))
    else
        U * sin(-μ*t + ψ + (mapreduce(x->x[2]*cos(x[1]*ϕ + x[3]), +, eventuples)
                + mapreduce(x->x[2]*sin(x[1]*ϕ +x[3]), +, oddtuples)))
        #U * sin(-μ*t + ψ + Peven * cos(neven * ϕ  + rotationeven) + Podd * sin(nodd * ϕ  + rotationodd))
    end
end

function odf_time_evolution(ρ::Float64, ϕ::Float64, zernikeeven::Array, zernikeodd::Array,
                            Γ::Float64=1/62, evolution_time::Float64=50.E-6;
                            ω::Float64=2*π*43.8E3, μ::Float64=0.0, optimize=true, bessel=false, small=false)
    b = SpinBasis(1//2)
    ψ0 = 1/sqrt(2) * (spindown(b) + spinup(b))
    T = [0:0.01E-6:evolution_time;]
    H(t, ψ) = H_odf(ρ, ϕ, t, zernikeeven, zernikeodd, μ=μ, ω=ω, bessel=bessel, small=small)*sigmaz(b), [sigmam(b)], [dagger(sigmam(b))]
    tout, ρ = timeevolution.master_dynamic(T, ψ0, H; rates=[Γ])
    trace = Real.(expect(sigmax(b), ρ))
    if optimize
        minimum(trace), T[argmin(trace)]
    else
        trace
    end
end
```

# Both Zernike [0,1]

```julia
zernike = [0, 1]
Γ = (1/62) 
rates = [Γ]
ρ = 1E-6
ϕ = 0.
b = SpinBasis(1//2)
ω = 2 * π * 43.8E3
μ = (length(zernike) - 1) * ω
ψ0 = 1/sqrt(2) * (spindown(b) + spinup(b))
```

```julia
# Note that between notebooks we moved to used seconds, rather than microseconds.
max_simulation_time = 50.E-6
pop, max_simulation_time = odf_time_evolution(ρ, ϕ, zernike, zernike, Γ, max_simulation_time, small=true, bessel=true, optimize=true)

H(t, _) = H_odf(ρ, ϕ, t, zernike, zernike, small=true, bessel=true) * sigmaz(b), [sigmaz(b)], [dagger(sigmaz(b))]
T = [0:0.01E-6:2*max_simulation_time;]
tout, ψ = timeevolution.master_dynamic(T, ψ0, H; rates=rates)

ω = 2 * π * 43.8E3
μ = (length(zernike) - 1) * ω
# Measure in the x basis
excitation = Real.(expect(sigmax(b), ψ))
large_bessel = odf_time_evolution(ρ, ϕ, zernike, zernike, Γ, 2*max_simulation_time; ω=ω, μ=μ, optimize=false, bessel=true, small=false)
exact = odf_time_evolution(ρ, ϕ, zernike, zernike, Γ, 2*max_simulation_time; ω=ω, μ=μ, optimize=false, bessel=false, small=false);
```

```julia
using Plots
plot(T, [excitation, large_bessel, exact])
```

# Multiple Zernike Orders

```julia
zernike1 = [0, 1]
μ1 = (length(zernike1) - 1) * ω
zernike2 = [0, 0, 1]
μ2 = (length(zernike2) - 1) * ω
```

# We can split the evolution time and find a 2π pulse under the approximate dynamics to compare against

```julia
max_simulation_time = 100.E-6
T = [0:max_simulation_time/1000.:max_simulation_time;]
excitations = []
for time in T
    function H(t, _)
        if t < time/2
            H_odf(ρ, ϕ, t, [0], zernike1, small=true, bessel=true) * sigmaz(b), [sigmaz(b)], [dagger(sigmaz(b))]
        else
            H_odf(ρ, ϕ, t, [0], zernike2, small=true, bessel=true) * sigmaz(b), [sigmaz(b)], [dagger(sigmaz(b))]
        end
    end
    if time == 0
        continue
    else
        t = [0:time/1000.:time;]
    end
    tout, ψ = timeevolution.master_dynamic(t, ψ0, H; rates=rates)
    push!(excitations, last(Real.(expect(sigmax(b), ψ))))
end
```

```julia
using PyPlot
plot(T[2:length(T)], excitations)
```

# 2π time under bessel approximation, simultaneously with two orders

```julia
T = [0:0.01E-6:max_simulation_time;]
# If I only have even terms it seems like it doesn't reach a minimum very quickly...
function H(t, _)
        if t < max_simulation_time/2
            H_odf(ρ, ϕ, t, [0], zernike1, small=true, bessel=true) * sigmaz(b), [sigmaz(b)], [dagger(sigmaz(b))]
        else
            H_odf(ρ, ϕ, t, [0], zernike2, small=true, bessel=true) * sigmaz(b), [sigmaz(b)], [dagger(sigmaz(b))]
        end
end
_, ψTARGET = timeevolution.master_dynamic(T, ψ0, H; rates=[Γ]);
ψTARGET = last(ψTARGET)
```

# Fidelity under exact dynamics, sequential application

```julia
H(t, _) = H_odf(ρ, ϕ, t, [0], [zernike1],  μ=μ1, ω=ω, bessel=false, small=false)*sigmaz(b), [sigmam(b)], [dagger(sigmam(b))]
_, ψ1SEQ = timeevolution.master_dynamic(T, ψ0, H; rates=[Γ])
ψ1SEQ = last(ψ1SEQ);
H(t, _) = H_odf(ρ, ϕ, t, [0], [zernike2], μ=μ2, ω=ω, bessel=false, small=false)*sigmaz(b), [sigmam(b)], [dagger(sigmam(b))]
_, ψ2SEQ = timeevolution.master_dynamic(T, ψ1SEQ, H; rates=[Γ]);
ψ2SEQ = last(ψ2SEQ);
```

```julia
fidelity(ψTARGET.data, ψ2SEQ.data)
```
